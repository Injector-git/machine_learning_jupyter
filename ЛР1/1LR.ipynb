15. Вычислите для этого столбца среднее значение, моду, медиану, дис

```python
print("Среднее:")
print(df['trestbps'].mean())
print("Мода:")
print(df['trestbps'].mode()[0])
print("Медиана:")
print(df['trestbps'].median())
print("Дисперсия:")
print(df['trestbps'].var())
print("Стандартное отклонение:")
print(df['trestbps'].std())
```
